```python
def categorize_time(time_value):
    if pd.Timestamp('00:00').time() <= time_value < pd.Timestamp('06:00').time():
        return 'Early Morning (00:00 - 06:00)'
    elif pd.Timestamp('06:00').time() <= time_value < pd.Timestamp('12:00').time():
        return 'Morning (06:00 - 12:00)'
    elif pd.Timestamp('12:00').time() <= time_value < pd.Timestamp('15:00').time():
        return 'Afternoon (12:00 - 15:00)'
    elif pd.Timestamp('15:00').time() <= time_value < pd.Timestamp('18:00').time():
        return 'Evening (15:00 - 18:00)'
    else:
        return 'Night (18:00 - 00:00)'
```

```python
df
```

```python
df['Time Category'] = df['Departure Time'].apply(categorize_time)
```

```python
df.info()
```

```python
flight_count = df.groupby(['Time Category']).size().reset_index().rename(columns = {0:'Flight count'})
flight_count
```

```python
def categorize_dis(amt):
    if amt < 200:
        return 'Low Discount'
    elif 200<= amt < 400:
        return 'Medium Discount'
    else:
        return 'High Discount'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\ARJUN\Downloads\flight_info - flight_info.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Promo-code'].apply(lambda x:x.split("'")[3])
```

```python
df['Discount'] = df['Promo-code'].apply(lambda x:x.split("'")[1])
df['Coupon code'] = df['Promo-code'].apply(lambda x:x.split("'")[3])
```

```python
df.head(2)
```

```python
df.drop(['Promo-code'],axis = 1,inplace = True)
```

```python
df.head(2)
```

```python
df['Departure Hour'] = df['Departure Time'].apply(lambda x:x.split(":")[0])
df['Destination Hour'] = df['Destination Time'].apply(lambda x:x.split(":")[0])
```

```python
df.head()
```

```python
df.info()
```

```python
df['Discount'] = df['Discount'].astype(int)

```

```python
df['Destination Hour'] = df['Destination Hour'].astype(int)
df['Departure Hour'] = df['Departure Hour'].astype(int)
```

```python
df.info()
```

```python
def categorize_dis(amt):
    if amt < 200:
        return 'Low Discount'
    elif 200<= amt < 400:
        return 'Medium Discount'
    else:
        return 'High Discount'
```

```python
df['Dis Category'] = df['Discount'].apply(categorize_dis)
```

```python
df
```

```python
df['Flight Name'] = df['Flight Name'].apply(lambda x:x.split("\n")[0])
```

```python
df['Flight Name'].unique()
```

```python
df['Flight Name'].value_counts()[['SG 8269"','FLY91']]
```

```python
df
```

```python
df.info()
```

```python
df['Departure Time'] = pd.to_datetime(df['Departure Time'],format = '%H:%M', exact = True)
```

```python
df['Destination Time'] = pd.to_datetime(df['Destination Time'],format = '%H:%M', exact = True)
```

```python
df.info()
```

```python

sns.relplot(data=df, x="Departure Time", y="Discount", hue="Dis Category" , style ='Flight Name',size = 'Flight Name')
```

```python
df = df.drop(df[~df['Flight Name'].isin(['Vistara', 'Air India','SpiceJet','IndiGo'])].index)
```

```python
df['Flight Name'].unique()
```

```python
df['Departure Time'].nunique()
```

```python
fig = sns.relplot(data=df, x="Departure Time", y="Destination Time", hue="Dis Category" , style ='Flight Name',size = "Flight Name")
fig.set_axis_labels('Departure Time', 'Destination Time')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Price1'] = df['Price'].apply(lambda x:x.split(',')[0]).str.split(' ').str[1]
```

```python
df['Price2'] = df['Price'].apply(lambda x:x.split(',')[1])
```

```python
df['price'] = df['Price1'] + df['Price2']
```

```python
df.head()
```

```python
df.drop(columns=['Price','Price1','Price2'],axis = 1,inplace = True)
```

```python
df['price'] = df['price'].astype(int)
```

```python
df.info()
```

```python
fig = sns.relplot(data=df, x="Departure Time", y="price", hue="Dis Category" , style ='Flight Name',size = "Flight Name")
fig.set_axis_labels('Departure Time', 'price')
```

```python
df.head()
```

```python
custom_palette = {
    'High Discount': 'red',
    'Medium Discount': 'blue',
    'Low Discount': 'green'
}

plt.figure(figsize=(12, 6))
sns.scatterplot(
    x = 'Departure Time',
    y = 'price',
    hue='Dis Category',
    style = 'Flight Name',
    size ='Flight Name',
    palette=custom_palette,
    data = df
)

plt.title("Scatter Plot of Departure Hour vs price")
plt.xlabel("Departure Time")
plt.ylabel("price")
plt.grid(True)
plt.show()
```
